We can convert python to JSON and this is called Serializing

```python
import json
a={"name":"John","age":"21","Salary":"21000"}
b=json.dumps(a)
print(b)
```

note : the data of json mostly looks like a dictionary

And we can convert json to python DeSerializing

```python
with open("Sample.json","r") as read_it:
    data=json.load(read_it)
```

```python
from datetime import datetime

def convert_date_format(date_str):
    try:
        # Try format with milliseconds
        return datetime.strptime(date_str, '%Y-%m-%dT%H:%M:%S.%fZ').strftime('%m/%d/%Y')
    except ValueError:
        try:
            # Try format without milliseconds
            return datetime.strptime(date_str, '%Y-%m-%dT%H:%M:%SZ').strftime('%m/%d/%Y')
        except ValueError:
            # If already in MM/DD/YYYY format, return as-is
            return date_str

# Apply the function to the Date column
df['newDate'] = df['Date'].apply(convert_date_format)

print(df)
```
